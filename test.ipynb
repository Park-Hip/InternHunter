```python
def extract_job_sections(row):
    text = str(row['description'])

    # Define the headers we are looking for
    headers = {
        'desc': r"Mô tả công việc",
        'req': r"Yêu cầu ứng viên",
        'ben': r"Quyền lợi",
        'loc': r"Địa điểm"
    }

    pos = {}
    for key, pattern in headers.items():
        match = re.search(pattern, text)
        pos[key] = (match.start(), match.end()) if match else (None, None)

    def get_text_between(start_idx, end_idx):
        if start_idx is None: return ""
        return text[start_idx:end_idx].strip()

    # logic: [End of Header A] to [Start of Header B]
    row['job_description_clean'] = get_text_between(pos['desc'][1], pos['req'][0])
    row['requirement'] = get_text_between(pos['req'][1], pos['ben'][0])
    row['benefit'] = get_text_between(pos['ben'][1], pos['loc'][0])

    return row
```

```python
df = df.apply(extract_job_sections, axis=1)
```
