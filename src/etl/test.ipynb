```python
import sqlite3
from src.settings.settings import DATA_DIR

db_path = (DATA_DIR / "db" / "jobs.db").resolve()
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# The timestamps you want to target
# Make sure these match EXACTLY what is in the DB
timestamps_to_delete = [
    "2026-02-03T14:25:29.441760",
    "2026-02-03T14:25:43.317473",
    "2026-02-03T14:25:56.365549"
]

try:
    # 1. Create dynamic placeholders (?, ?, ?)
    placeholders = ','.join(['?'] * len(timestamps_to_delete))
    
    # 2. Build the query safely
    # Note: 'scraped_at' is the standard column name we defined earlier.
    # If you really renamed it to 'is_scraped', change it back here.
    query = f"DELETE FROM raw_jobs WHERE scraped_at IN ({placeholders})"
    
    # 3. Execute with the list as arguments
    cursor.execute(query, timestamps_to_delete)
    
    conn.commit()
    print(f"✅ Deleted {cursor.rowcount} rows matching the timestamps.")

except Exception as e:
    print(f"❌ Error: {e}")
finally:
    conn.close()
```
